```python
import tensorflow as tf
import keras_hub
from keras import layers
```

```python
# This will print out every valid preset string for ResNetBackbone
print(keras_hub.models.ResNetBackbone.presets.keys())
```

```python
df = keras_hub.models.ResNetBackbone.from_preset("resnet_18_imagenet", load_weights=True) #resnetbackbone -> not final layer
# from preset means same depth, sizes and settings, but diff weight bcz the dataset is diff
```

```python
inputs = tf.keras.Input((32,32,3))
```

```python
x = layers.Resizing(244,244)(inputs)
```

```python
x = df(x)
```

```python
x = layers.GlobalAveragePooling1D()(x)
```
