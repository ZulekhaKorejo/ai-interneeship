```python
import numpy as np
```

```python
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])
array1 = np.array([
    [ 2 , 3],
    [ 3 , 5]
])
array2 = np.array([
     [ 1 , 0],
     [ 4 , 5]])
```

```python
# for 2x2 dot multiplication between matrices we use the dot mutiplication (@) which results in the muliplication of arr1(rows) x arr2(columns)
# in the prev example it'll be array1 @ array2=
# (1,1) = (2*1) + (3*4) = 2+12 = 14
# (1,2) = (2*0) + (3*5) = 0+15 = 15
# (2,1) = (3*1) + (5*4) = 3+20 = 23
# (2,2) = (3*0) + (5*5) = 0=25 = 25
# for confirmation:
arr_multi= array1 @ array2
print (arr_multi)
```

```python
# 15- Dot product:
# vector dot muliplication is easier but they should be equal in length
# for example vector1 @ vector 2 =
# (1*4) + (2*5) + (3*6) = 32
# this helpps in finding how much do vectors/arrays align or agree with each other
# which helps the nueral networks (core of Deep learning) to make decisions, for example answering questions like "would this be a good ad for tis person?"
vec_multi= vector1 @ vector2
print(vec_multi)
```
