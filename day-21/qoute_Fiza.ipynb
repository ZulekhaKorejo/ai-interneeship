```python
import pandas as pd
import numpy as np

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, Dense

from tensorflow.keras.utils import to_categorical
```

```python
df = pd.read_csv("qoute_dataset.csv")

df.head()
```

```python
df.columns
```

```python
text = " ".join(df['quote'].astype(str))

print(text[:500])
```

```python
sentences = text.lower().split(".")
```

```python
tokenizer = Tokenizer()

tokenizer.fit_on_texts(sentences)

total_words = len(tokenizer.word_index) + 1

print(tokenizer.word_index)
```

```python
input_sequences = []

for sentence in sentences:

    words = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1, len(words)):

        input_sequences.append(words[:i+1])


print(input_sequences[:5])
```

```python
max_length = max([len(x) for x in input_sequences])

input_sequences = pad_sequences(
    input_sequences,
    maxlen=max_length,
    padding='pre'
)

input_sequences[:5]
```

```python
X = input_sequences[:, :-1]

y = input_sequences[:, -1]
```

```python
y = to_categorical(
    y,
    num_classes=total_words
)
```

```python
rnn_model = Sequential()


rnn_model.add(
    Embedding(
        total_words,
        100,
        input_length=X.shape[1]
    )
)


rnn_model.add(
    SimpleRNN(100)
)


rnn_model.add(
    Dense(total_words, activation='softmax')
)



rnn_model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)


rnn_model.summary()
```

```python
rnn_model.fit(
    X,
    y,
    epochs=50,
    verbose=1
)
```

```python
rnn_model.save('rnn_text_generator_model.keras')
```

```python
lstm_model = Sequential()


lstm_model.add(
    Embedding(
        total_words,
        100,
        input_length=X.shape[1]
    )
)


lstm_model.add(
    LSTM(100)
)


lstm_model.add(
    Dense(total_words, activation='softmax')
)


lstm_model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)


lstm_model.summary()
```

```python
lstm_model.fit(
    X,
    y,
    epochs=100,
    verbose=1
)
```

```python
lstm_model.save('lstm_text_generator_model.keras')
```

```python
def predict_next_word(model, text):

    for i in range(5):

        token_list = tokenizer.texts_to_sequences([text])[0]

        token_list = pad_sequences(
            [token_list],
            maxlen=max_length-1,
            padding='pre'
        )


        predicted = model.predict(token_list, verbose=0)

        predicted_word_index = np.argmax(predicted)


        output_word = ""

        for word,index in tokenizer.word_index.items():

            if index == predicted_word_index:
                output_word = word
                break


        text = text + " " + output_word


    return text
```
