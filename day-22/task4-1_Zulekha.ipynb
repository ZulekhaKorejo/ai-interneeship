```python
from tiktoken._educational import SimpleBytePairEncoding, train_simple_encoding
```

```python
encoder = train_simple_encoding()   #holds the created vocabulary
```

```python
# create variables of the two languges you wanna compare

English = 'AI is transforming software.'
Arabic = "يُحدث الذكاء الاصطناعي ثورة جذرية في قطاع البرمجيات"
long = "antidisestablishmentarianism"
```

```python
# encodes them to numbers

eng_encoded = encoder.encode(English)
arb_encoded = encoder.encode(Arabic)
long_encoded = encoder.encode(long)
```

```python
# checking
print(f"Sentence: {Arabic} \n words: {len(Arabic.split())} \n token ID: {arb_encoded} \n tokens: {len(arb_encoded)} \n ---------------------\n \
      Sentence: {English} \n words: {len(English.split())} \n token ID: {eng_encoded} \n tokens: {len(eng_encoded)} \n---------------------\n \
      Sentence: {long} \n words: {len(long.split())} \n token ID: {long_encoded} \n tokens: {len(long_encoded)} \n ---------------------")
```

```python
# words are less than tokens, in all of them even with punctiations counted alone
#  but tokens are less than letters (except in arabic), so it cuts off words and counts it insted of counting letters or words
#  as it shows the non latin languages like arabic make much more tokens than english
#  also the long words are broken into many smaller parts and tokenized instead of taking the whole word
```

```python
# showing how it was tokenized
eng_byte = encoder.decode_tokens_bytes(eng_encoded)
arb_byte = encoder.decode_tokens_bytes(arb_encoded)
long_byte = encoder.decode_tokens_bytes(long_encoded)
eng_byte
```

```python
# as u see it has a prefix for each token i wanna remove that for readability using b.decode(utf-8)
eng_clean = [b.decode('utf-8', errors='replace') for b in eng_byte] #loop oh all the b in the list
long_clean = [b.decode('utf-8', errors='replace') for b in long_byte]
arb_clean = [f"\\x{b[0]:02x}" for b in arb_byte] # b[0] returns the int value of each, :02x for formatting in two dgits and lower case
```

```python
print(f"Sentence: {Arabic} \n words: {len(Arabic.split())} \n tokenized to: {arb_clean} \n tokens: {len(arb_clean)} \n ---------------------\n \
      Sentence: {English} \n words: {len(English.split())} \n tokenized to: {eng_clean} \n tokens: {len(eng_clean)} \n ---------------------\n \
        Sentence: {long} \n words: {len(long.split())} \n tokenized to: {long_clean} \n tokens: {len(long_clean)} \n ---------------------\n")
```
